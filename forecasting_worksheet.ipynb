```python
##Adding year, month, day columns to dataframe
load_daily['year']=load_daily.index.year
load_daily['month']=load_daily.index.month
load_daily['weekday']=load_daily.index.day_name()
load_daily.head(10)
```

```python
load_7d = load_daily.rolling(7, center=True).mean()
load_7d.head(10)
```

```python
# Start and end of the date range to extract
start, end = '2018-01', '2018-06'
# Plot daily, weekly resampled, and 7-day rolling mean time series together
fig, ax = plot.subplots()
ax.plot(load_daily.loc[start:end, 'ontario_demand'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(load_7d.loc[start:end, 'ontario_demand'],
marker='.', linestyle='-', label='7-d Rolling Mean')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

```python
# The min_periods=360 argument accounts for a few isolated missing days in the
# wind and solar production time series
load_365d = load_daily.rolling(window=365, center=True, min_periods=360).mean()
```

```python
# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plot.subplots()
ax.plot(load_daily['ontario_demand'], marker='.', markersize=2, color='0.6',
linestyle='None', label='Daily')
ax.plot(load_7d['ontario_demand'], linewidth=2, label='7-d Rolling Mean')
ax.plot(load_365d['ontario_demand'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption');
```
