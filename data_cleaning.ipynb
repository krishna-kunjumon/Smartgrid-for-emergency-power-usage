# Missing Values

```python
missing_value_count=forecast.isnull().sum()
```

```python
# how many total missing values do we have?
total_cells = np.product(forecast.shape)
total_missing = missing_value_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

More than half of the cells in the dataset are empty. Lets look into the data to determine why the data is empty. The process of filling missing data is called imputation.

```python
forecast=forecast.drop(['temp_max_flag','temp_min_flag','temp_mean_flag','heat_deg_days_flag','cool_deg_days_flag','total_precip_mm_flag','iso_3166_2_code','census_fips_code','metro_area','sub_region_2','sub_region_1','country_region','country_region_code'], axis=1)
```

```python
missing_value_count=forecast.isnull().sum()
```

```python
missing_value_count
```

```python
forecast.shape
```

```python
forecast[forecast['market_demand'].isnull()==True]
forecast[forecast['ontario_demand'].isnull()==True]
```

```python
values = {'market_demand': 0, 'ontario_demand': 0}
forecast=forecast.fillna(value=values)
```

```python
forecast['temp_mean']=forecast[['temp_max','temp_min']].mean(axis=1)
```

```python
forecast[forecast['driving'].isnull()==True]
```

```python
values = {'driving': 0, 'walking': 0, 'transit':0, 'residential_percent_change_from_baseline':0,'workplaces_percent_change_from_baseline':0,'transit_stations_percent_change_from_baseline':0,'parks_percent_change_from_baseline':0,'grocery_and_pharmacy_percent_change_from_baseline':0,'retail_and_recreation_percent_change_from_baseline':0}
forecast=forecast.fillna(value=values)
```

```python
forecast=forecast[forecast.index<'2021-06-20']
```

```python
forecast=forecast.drop(['spd_of_max_gust_kmh_flag','spd_of_max_gust_kmh','dir_of_max_gust_10sdeg_flag','dir_of_max_gust_10sdeg','total_rain_mm','total_rain_mm_flag','total_snow_cm','total_snow_cm_flag','snow_on_ground_cm','snow_on_ground_cm_flag'], axis=1)
```
